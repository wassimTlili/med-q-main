```python
import pdfplumber
import re
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

class MedicalRAGSystem:
    def __init__(self, pdf_path, llm):
        self.pdf_path = pdf_path
        self.llm = llm
        self.embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        self.qa_chain = None
        self.setup_rag()

    def extract_pdf_with_pages(self):
        """Extract PDF text while preserving page numbers and structure"""
        pages_content = []

        with pdfplumber.open(self.pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages, 1):
                text = page.extract_text()
                if text:
                    pages_content.append({
                        'page_number': page_num,
                        'content': text,
                        'clean_content': self.clean_medical_text(text)
                    })

        return pages_content

    def clean_medical_text(self, text):
        """Clean text while preserving medical terms and structure"""
        # Preserve important punctuation for medical context
        # Keep dots, commas, parentheses, hyphens
        text = re.sub(r'[^\w\s\.\,\(\)\-\:\;]', ' ', text)

        # Remove excessive whitespace but preserve paragraph breaks
        text = re.sub(r'\n\s*\n', '\n\n', text)
        text = re.sub(r' +', ' ', text)

        return text.strip()

    def create_structured_chunks(self, pages_content):
        """Create chunks that preserve medical structure with headings"""
        docs = []

        # Custom splitter for medical content
        splitter = RecursiveCharacterTextSplitter(
            chunk_size=800,
            chunk_overlap=150,
            length_function=len,
            separators=["\n\n", "\n", ". ", " "]  # Respect medical structure
        )

        for page_data in pages_content:
            page_num = page_data['page_number']
            content = page_data['clean_content']

            # Split content while trying to keep headings with their content
            chunks = splitter.split_text(content)

            for i, chunk in enumerate(chunks):
                # Create metadata-rich document
                doc = Document(
                    page_content=chunk,
                    metadata={
                        'page': page_num,
                        'chunk_id': f"page_{page_num}_chunk_{i+1}",
                        'source': f"Page {page_num}"
                    }
                )
                docs.append(doc)

        return docs

    def setup_rag(self):
        """Setup the RAG system"""
        print("Extracting PDF content with page tracking...")
        pages_content = self.extract_pdf_with_pages()

        print("Creating structured chunks...")
        docs = self.create_structured_chunks(pages_content)

        print("Building vector store...")
        vectorstore = FAISS.from_documents(docs, self.embedding_model)

        # Create retriever
        retriever = vectorstore.as_retriever(
            search_type="similarity",
            search_kwargs={"k": 8}  # Optimized for focused retrieval
        )

        # Enhanced medical prompt
        mcq_prompt = """Vous êtes un expert médical analysant des questions à choix multiples. Utilisez UNIQUEMENT le texte médical fourni pour évaluer chaque option.

Contexte médical:
{context}

Question: {question}

Instructions:
1. Pour CHAQUE option (A, B, C, D, E), déterminez si elle est VRAIE ou FAUSSE basé UNIQUEMENT sur le contexte fourni
2. Fournissez une justification spécifique du texte pour chaque option
3. Citez les phrases exactes du contexte qui soutiennent votre analyse
4. Si l'information n'est pas disponible, indiquez "Non mentionné dans le texte fourni"
5. À la fin, listez les options CORRECTES

Format de réponse:
Option A: [VRAI/FAUX] - [Justification avec citations exactes]
Option B: [VRAI/FAUX] - [Justification avec citations exactes]
Option C: [VRAI/FAUX] - [Justification avec citations exactes]
Option D: [VRAI/FAUX] - [Justification avec citations exactes]
Option E: [VRAI/FAUX] - [Justification avec citations exactes]

RÉPONSES CORRECTES: [Lister les options correctes]

Analyse:"""

        PROMPT = PromptTemplate(
            template=mcq_prompt,
            input_variables=["context", "question"]
        )

        # Create QA chain
        self.qa_chain = RetrievalQA.from_chain_type(
            llm=self.llm,
            chain_type="stuff",
            retriever=retriever,
            return_source_documents=True,
            chain_type_kwargs={"prompt": PROMPT}
        )

        print("RAG system ready!")

    def analyze_single_question(self, question_text):
        """Analyze a single MCQ question"""
        result = self.qa_chain(question_text)

        print("="*80)
        print("ANALYSE MCQ MÉDICAL")
        print("="*80)
        print(f"\nQuestion:\n{question_text}")
        print("\n" + "-"*60)
        print("ANALYSE BASÉE SUR LE TEXTE MÉDICAL:")
        print("-"*60)
        print(result['result'])

        print("\n" + "="*60)
        print("RÉFÉRENCES SOURCES AVEC PAGES:")
        print("="*60)

        for i, doc in enumerate(result['source_documents'], 1):
            page_info = doc.metadata.get('source', 'Page inconnue')
            print(f"\n[Source {i} - {page_info}]")
            print(f"Contenu: {doc.page_content}")
            print("-" * 50)

        return result

    def analyze_multiple_questions(self, questions_list):
        """Analyze multiple questions independently"""
        results = []

        for i, question in enumerate(questions_list, 1):
            print(f"\n{'='*20} QUESTION {i} {'='*20}")
            result = self.analyze_single_question(question)
            results.append(result)
            print(f"\n{'='*50}\n")

        return results

# Usage Example
```

```python
import pdfplumber
import re
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

class IndependentChoiceRAG:
    def __init__(self, pdf_path, llm):
        self.pdf_path = pdf_path
        self.llm = llm
        self.embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        self.vectorstore = None
        self.setup_rag()

    def extract_pdf_with_pages(self):
        """Extract PDF text while preserving page numbers"""
        pages_content = []

        with pdfplumber.open(self.pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages, 1):
                text = page.extract_text()
                if text:
                    pages_content.append({
                        'page_number': page_num,
                        'content': text,
                        'clean_content': self.clean_medical_text(text)
                    })

        return pages_content

    def clean_medical_text(self, text):
        """Clean text while preserving medical terms"""
        text = re.sub(r'[^\w\s\.\,\(\)\-\:\;]', ' ', text)
        text = re.sub(r'\n\s*\n', '\n\n', text)
        text = re.sub(r' +', ' ', text)
        return text.strip()

    def create_structured_chunks(self, pages_content):
        """Create chunks with page metadata"""
        docs = []

        splitter = RecursiveCharacterTextSplitter(
            chunk_size=800,
            chunk_overlap=300,
            length_function=len,
            separators=["\n\n", "\n", ". ", " "]
        )

        for page_data in pages_content:
            page_num = page_data['page_number']
            content = page_data['clean_content']

            chunks = splitter.split_text(content)

            for i, chunk in enumerate(chunks):
                doc = Document(
                    page_content=chunk,
                    metadata={
                        'page': page_num,
                        'chunk_id': f"page_{page_num}_chunk_{i+1}",
                        'source': f"Page {page_num}"
                    }
                )
                docs.append(doc)

        return docs

    def setup_rag(self):
        """Setup the RAG system"""
        print("Setting up RAG system...")
        pages_content = self.extract_pdf_with_pages()
        docs = self.create_structured_chunks(pages_content)

        self.vectorstore = FAISS.from_documents(docs, self.embedding_model)
        print("RAG system ready!")

    def search_for_choice(self, choice_text, k=10):
        """Search specifically for a single choice"""
        retriever = self.vectorstore.as_retriever(
            search_type="similarity",
            search_kwargs={"k": k}
        )

        # Search for relevant documents
        docs = retriever.get_relevant_documents(choice_text)
        return docs

    def analyze_single_choice(self, choice_letter, choice_text, context_question=""):
        """Analyze a single choice independently"""

        # Search for this specific choice
        relevant_docs = self.search_for_choice(choice_text)

        # Create context from retrieved documents
        context = "\n\n".join([doc.page_content for doc in relevant_docs])

        # Prompt for single choice analysis
        single_choice_prompt = f"""Vous êtes un expert médical. Analysez cette affirmation médicale en utilisant UNIQUEMENT le contexte fourni.

Contexte médical:
{context}

Question générale: {context_question}

Affirmation à analyser:
{choice_letter}. {choice_text}

Instructions:
1. Déterminez si cette affirmation est VRAIE ou FAUSSE basé UNIQUEMENT sur le contexte fourni
2. Fournissez une justification détaillée avec citations exactes du texte
3. Si l'information n'est pas dans le contexte, indiquez "Non mentionné dans le texte fourni"

Répondez au format:
CHOIX {choice_letter}: [VRAI/FAUX]
JUSTIFICATION: [Explication détaillée avec citations exactes du texte]
"""

        # Create prompt template and chain for single choice
        prompt_template = PromptTemplate(
            template=single_choice_prompt,
            input_variables=[]
        )

        # Use LLM with the prompt
        try:
            response = self.llm.invoke(single_choice_prompt)
            if hasattr(response, 'content'):
                analysis_result = response.content
            else:
                analysis_result = str(response)
        except Exception as e:
            analysis_result = f"Erreur d'analyse: {str(e)}"

        return {
            'choice': choice_letter,
            'text': choice_text,
            'analysis': analysis_result,
            'sources': relevant_docs
        }

    def parse_mcq_question(self, question_text):
        """Parse MCQ question to extract choices"""
        lines = question_text.strip().split('\n')

        question_title = ""
        choices = {}

        for line in lines:
            line = line.strip()
            if line.startswith(('A.', 'B.', 'C.', 'D.', 'E.')):
                choice_letter = line[0]
                choice_text = line[2:].strip()
                choices[choice_letter] = choice_text
            elif not line.startswith(('A.', 'B.', 'C.', 'D.', 'E.')) and line:
                question_title += line + " "

        return question_title.strip(), choices

    def analyze_mcq_independently(self, question_text):
        """Analyze each MCQ choice independently"""

        # Parse the question
        question_title, choices = self.parse_mcq_question(question_text)

        print("="*80)
        print("ANALYSE INDÉPENDANTE DES CHOIX MCQ")
        print("="*80)
        print(f"Question: {question_title}")
        print("="*80)

        results = {}
        correct_choices = []

        # Analyze each choice independently
        for choice_letter, choice_text in choices.items():
            print(f"\n{'='*20} ANALYSE DU CHOIX {choice_letter} {'='*20}")
            print(f"Affirmation: {choice_text}")
            print("-"*60)

            # Independent analysis for this choice
            result = self.analyze_single_choice(choice_letter, choice_text, question_title)

            print("RÉSULTAT DE L'ANALYSE:")
            print(result['analysis'])

            print(f"\nSOURCES UTILISÉES POUR LE CHOIX {choice_letter}:")
            print("-"*40)
            for i, doc in enumerate(result['sources'], 1):
                page_info = doc.metadata.get('source', 'Page inconnue')
                print(f"[Source {i} - {page_info}]")
                print(f"Extrait: {doc.page_content[:200]}...")
                print("-"*30)

            results[choice_letter] = result

            # Try to extract if it's correct (simple parsing)
            if "VRAI" in result['analysis'].upper():
                correct_choices.append(choice_letter)

        # Summary
        print("\n" + "="*80)
        print("RÉSUMÉ FINAL")
        print("="*80)
        print(f"Choix analysés: {', '.join(choices.keys())}")
        print(f"Choix probablement corrects: {', '.join(correct_choices) if correct_choices else 'Aucun détecté automatiquement'}")

        return results

    def analyze_multiple_mcq(self, questions_list):
        """Analyze multiple MCQ questions"""
        all_results = []

        for i, question in enumerate(questions_list, 1):
            print(f"\n{'#'*100}")
            print(f"{'#'*40} QUESTION {i} {'#'*40}")
            print(f"{'#'*100}")

            result = self.analyze_mcq_independently(question)
            all_results.append(result)

        return all_results

# Usage Example
def main(llm):  # Pass LLM as parameter
    # Initialize system
    rag_system = IndependentChoiceRAG("Gastrologie entérologie-23-36.pdf", llm)

    # Example question
    question1 = '''
    Question n° 1 (Tunis 2024, [niveau3]): Concernant les caractéristiques macroscopiques de l'ulcère chronique gastro-duodénal :
A. L'atteinte préférentielle est la petite courbure
B. Peut être multiple dans le syndrome de Zollinger Ellison
C. Présente un aspect de perte de substance linéaire
D. Le fond est recouvert de kératine
E. Son cratère est entouré de bourrelet boursouflé





    '''

    # Analyze with independent choice analysis
    results = rag_system.analyze_mcq_independently(question1)

    return results

# To use with your LLM:
results = main(llm)
```

```python
# Azure OpenAI Version - Matching Your System
import os
import requests
import json
import pdfplumber
import re
from typing import List, Dict, Any, Optional

class AzureOpenAIRAGSystem:
    def __init__(self, pdf_path: str):
        self.pdf_path = pdf_path
        self.chunks = []
        
        # Azure OpenAI configuration (matching your env)
        self.azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT", "").rstrip('/')
        self.api_key = os.getenv("AZURE_OPENAI_API_KEY", "")
        self.embedding_deployment = os.getenv("AZURE_OPENAI_EMBEDDING_DEPLOYMENT", "")
        self.chat_deployment = os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT", "")
        
        if not all([self.azure_endpoint, self.api_key, self.embedding_deployment]):
            raise ValueError("Missing Azure OpenAI configuration")
            
        print("Setting up Azure OpenAI RAG system...")
        self.setup_rag()

    def extract_pdf_with_pages(self):
        """Extract PDF text while preserving page numbers - exact notebook logic"""
        pages_content = []

        with pdfplumber.open(self.pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages, 1):
                text = page.extract_text()
                if text:
                    pages_content.append({
                        'page_number': page_num,
                        'content': text,
                        'clean_content': self.clean_medical_text(text)
                    })

        return pages_content

    def clean_medical_text(self, text):
        """Clean text while preserving medical terms - exact notebook logic"""
        # Preserve important punctuation for medical context
        text = re.sub(r'[^\w\s\.\,\(\)\-\:\;]', ' ', text)
        
        # Remove excessive whitespace but preserve paragraph breaks
        text = re.sub(r'\n\s*\n', '\n\n', text)
        text = re.sub(r' +', ' ', text)
        
        return text.strip()

    def create_structured_chunks(self, pages_content):
        """Create chunks - exact notebook logic with 800/300 overlap"""
        chunks = []
        
        # Match notebook chunking exactly
        chunk_size = 800
        chunk_overlap = 300
        separators = ["\n\n", "\n", ". ", " "]
        
        for page_data in pages_content:
            page_num = page_data['page_number']
            content = page_data['clean_content']
            
            # Simple recursive text splitter (mimic notebook behavior)
            page_chunks = self.recursive_split(content, chunk_size, chunk_overlap, separators)
            
            for i, chunk in enumerate(page_chunks):
                chunks.append({
                    'text': chunk,
                    'page': page_num,
                    'chunk_id': f"page_{page_num}_chunk_{i+1}",
                    'source': f"Page {page_num}"
                })
                
        return chunks

    def recursive_split(self, text: str, chunk_size: int, overlap: int, separators: List[str]) -> List[str]:
        """Simple recursive text splitter matching notebook logic"""
        if len(text) <= chunk_size:
            return [text] if text.strip() else []
        
        # Try each separator
        for sep in separators:
            if sep in text:
                parts = text.split(sep)
                if len(parts) > 1:
                    chunks = []
                    current_chunk = ""
                    
                    for part in parts:
                        if not part.strip():
                            continue
                            
                        # Add separator back except for space
                        part_with_sep = part + (sep if sep != " " else "")
                        
                        if len(current_chunk) + len(part_with_sep) <= chunk_size:
                            current_chunk += part_with_sep
                        else:
                            if current_chunk.strip():
                                chunks.append(current_chunk.strip())
                            current_chunk = part_with_sep
                    
                    if current_chunk.strip():
                        chunks.append(current_chunk.strip())
                    
                    # Add overlap
                    overlapped_chunks = []
                    for i, chunk in enumerate(chunks):
                        if i > 0 and overlap > 0:
                            # Add overlap from previous chunk
                            prev_chunk = chunks[i-1]
                            if len(prev_chunk) > overlap:
                                overlap_text = prev_chunk[-overlap:] + " " + chunk
                                overlapped_chunks.append(overlap_text)
                            else:
                                overlapped_chunks.append(chunk)
                        else:
                            overlapped_chunks.append(chunk)
                    
                    return overlapped_chunks
        
        # Fallback: character-based splitting
        chunks = []
        for i in range(0, len(text), chunk_size - overlap):
            chunk = text[i:i + chunk_size]
            if chunk.strip():
                chunks.append(chunk.strip())
        return chunks

    def embed_texts(self, texts: List[str]) -> List[List[float]]:
        """Embed texts using Azure OpenAI - matching your system"""
        url = f"{self.azure_endpoint}/openai/deployments/{self.embedding_deployment}/embeddings"
        headers = {
            "Content-Type": "application/json",
            "api-key": self.api_key
        }
        data = {
            "input": texts,
            "model": self.embedding_deployment
        }
        
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()
        
        result = response.json()
        return [item["embedding"] for item in result["data"]]

    def cosine_similarity(self, a: List[float], b: List[float]) -> float:
        """Calculate cosine similarity"""
        dot_product = sum(x * y for x, y in zip(a, b))
        norm_a = sum(x * x for x in a) ** 0.5
        norm_b = sum(x * x for x in b) ** 0.5
        
        if norm_a == 0 or norm_b == 0:
            return 0
        
        return dot_product / (norm_a * norm_b)

    def setup_rag(self):
        """Setup RAG system - notebook logic"""
        print("Extracting PDF content with page tracking...")
        pages_content = self.extract_pdf_with_pages()
        
        print("Creating structured chunks...")
        self.chunks = self.create_structured_chunks(pages_content)
        
        print(f"Created {len(self.chunks)} chunks")
        
        print("Embedding chunks...")
        chunk_texts = [chunk['text'] for chunk in self.chunks]
        
        # Batch embedding (like your system)
        batch_size = 64
        embeddings = []
        
        for i in range(0, len(chunk_texts), batch_size):
            batch = chunk_texts[i:i + batch_size]
            batch_embeddings = self.embed_texts(batch)
            embeddings.extend(batch_embeddings)
            print(f"Embedded batch {i//batch_size + 1}/{(len(chunk_texts) + batch_size - 1)//batch_size}")
        
        # Store embeddings in chunks
        for chunk, embedding in zip(self.chunks, embeddings):
            chunk['embedding'] = embedding
            
        print("RAG system ready!")

    def search_chunks(self, query: str, k: int = 10) -> List[Dict[str, Any]]:
        """Search chunks - notebook logic with k=10"""
        if not self.chunks:
            return []
            
        # Embed query
        query_embedding = self.embed_texts([query])[0]
        
        # Calculate similarities
        scored_chunks = []
        for chunk in self.chunks:
            score = self.cosine_similarity(query_embedding, chunk['embedding'])
            scored_chunks.append({
                'chunk': chunk,
                'score': score
            })
        
        # Sort by score and return top k
        scored_chunks.sort(key=lambda x: x['score'], reverse=True)
        return scored_chunks[:k]

    def analyze_single_choice(self, choice_letter: str, choice_text: str, context_question: str = "") -> Dict[str, Any]:
        """Analyze single choice - exact notebook logic"""
        
        # Search for this specific choice (notebook approach)
        relevant_chunks = self.search_chunks(choice_text, k=10)
        
        # Create context from retrieved documents
        context = "\n\n".join([result['chunk']['text'] for result in relevant_chunks])
        
        # Notebook prompt format
        prompt = f"""Vous êtes un expert médical. Analysez cette affirmation médicale en utilisant UNIQUEMENT le contexte fourni.

Contexte médical:
{context}

Question générale: {context_question}

Affirmation à analyser:
{choice_letter}. {choice_text}

Instructions:
1. Déterminez si cette affirmation est VRAIE ou FAUSSE basé UNIQUEMENT sur le contexte fourni
2. Fournissez une justification détaillée avec citations exactes du texte
3. Si l'information n'est pas dans le contexte, indiquez "Non mentionné dans le texte fourni"

Répondez au format:
CHOIX {choice_letter}: [VRAI/FAUX]
JUSTIFICATION: [Explication détaillée avec citations exactes du texte]
"""

        # Call Azure OpenAI Chat
        try:
            response = self.chat_completion(prompt)
            analysis_result = response
        except Exception as e:
            analysis_result = f"Erreur d'analyse: {str(e)}"

        return {
            'choice': choice_letter,
            'text': choice_text,
            'analysis': analysis_result,
            'sources': [result['chunk'] for result in relevant_chunks]
        }

    def chat_completion(self, prompt: str) -> str:
        """Azure OpenAI chat completion"""
        url = f"{self.azure_endpoint}/openai/deployments/{self.chat_deployment}/chat/completions"
        headers = {
            "Content-Type": "application/json",
            "api-key": self.api_key
        }
        data = {
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.6,
            "max_tokens": 1500
        }
        
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()
        
        result = response.json()
        return result["choices"][0]["message"]["content"]

    def parse_mcq_question(self, question_text: str):
        """Parse MCQ question - exact notebook logic"""
        lines = question_text.strip().split('\n')

        question_title = ""
        choices = {}

        for line in lines:
            line = line.strip()
            if line.startswith(('A.', 'B.', 'C.', 'D.', 'E.')):
                choice_letter = line[0]
                choice_text = line[2:].strip()
                choices[choice_letter] = choice_text
            elif not line.startswith(('A.', 'B.', 'C.', 'D.', 'E.')) and line:
                question_title += line + " "

        return question_title.strip(), choices

    def analyze_mcq_independently(self, question_text: str):
        """Analyze MCQ - exact notebook logic"""
        
        # Parse the question
        question_title, choices = self.parse_mcq_question(question_text)

        print("="*80)
        print("ANALYSE INDÉPENDANTE DES CHOIX MCQ")
        print("="*80)
        print(f"Question: {question_title}")
        print("="*80)

        results = {}
        correct_choices = []

        # Analyze each choice independently (notebook approach)
        for choice_letter, choice_text in choices.items():
            print(f"\n{'='*20} ANALYSE DU CHOIX {choice_letter} {'='*20}")
            print(f"Affirmation: {choice_text}")
            print("-"*60)

            # Independent analysis for this choice
            result = self.analyze_single_choice(choice_letter, choice_text, question_title)

            print("RÉSULTAT DE L'ANALYSE:")
            print(result['analysis'])

            print(f"\nSOURCES UTILISÉES POUR LE CHOIX {choice_letter}:")
            print("-"*40)
            for i, chunk in enumerate(result['sources'], 1):
                page_info = chunk.get('source', 'Page inconnue')
                print(f"[Source {i} - {page_info}]")
                print(f"Extrait: {chunk['text'][:200]}...")
                print("-"*30)

            results[choice_letter] = result

            # Try to extract if it's correct (simple parsing)
            if "VRAI" in result['analysis'].upper():
                correct_choices.append(choice_letter)

        # Summary
        print("\n" + "="*80)
        print("RÉSUMÉ FINAL")
        print("="*80)
        print(f"Choix analysés: {', '.join(choices.keys())}")
        print(f"Choix probablement corrects: {', '.join(correct_choices) if correct_choices else 'Aucun détecté automatiquement'}")

        return results

# Test with your Azure setup
def test_azure_rag():
    # Test if env vars are set
    required_vars = ["AZURE_OPENAI_ENDPOINT", "AZURE_OPENAI_API_KEY", "AZURE_OPENAI_EMBEDDING_DEPLOYMENT", "AZURE_OPENAI_CHAT_DEPLOYMENT"]
    missing = [var for var in required_vars if not os.getenv(var)]
    
    if missing:
        print(f"Missing environment variables: {missing}")
        print("Please set these before running the test")
        return None
    
    try:
        # Initialize with your PDF
        rag_system = AzureOpenAIRAGSystem("Gastrologie entérologie-23-36.pdf")
        
        # Test question
        question = '''
Question n° 1 (Tunis 2024, [niveau3]): Concernant les caractéristiques macroscopiques de l'ulcère chronique gastro-duodénal :
A. L'atteinte préférentielle est la petite courbure
B. Peut être multiple dans le syndrome de Zollinger Ellison
C. Présente un aspect de perte de substance linéaire
D. Le fond est recouvert de kératine
E. Son cratère est entouré de bourrelet boursouflé
        '''
        
        # Analyze with Azure OpenAI
        results = rag_system.analyze_mcq_independently(question)
        return results
        
    except Exception as e:
        print(f"Error: {e}")
        return None

# Uncomment to test
# test_azure_rag()
```
